```python
# PySpark Data Transformer
```

```python
order_schema = StructType([
    StructField("order_id",IntegerType(), True),
    StructField("customer_id",IntegerType(), True),
    StructField("product_id",IntegerType(), True),
    StructField("quantity",IntegerType(), True),
    StructField("price",DoubleType(), True),
    StructField("order_timestamp",StringType(), True),
    StructField("status",StringType(), True),
])
```

```python
prduct_schema = StructType([
    StructField("product_id",IntegerType(),True),
    StructField("product_name",StringType(),True),
    StructField("category",StringType(),True),
])
```

```python
orders_df = spark.read.schema(order_schema).option("header","true").csv("Files/Capstone_1/orders.csv")
```

```python
orders_df.show()
```

```python
product_df = spark.read.schema(prduct_schema).option("header","true").csv("Files/Capstone_1/products.csv")
```

```python
product_df.show()
```

```python
cust_df = spark.read.json("Files/Capstone_1/customers.json")
```

```python
cust_df.show()
```

```python
orders_df.filter(col("price").isNotNull()).show()
```

```python
orders_clean = orders_df.filter(col("price").isNotNull())
```

```python
orders_clean
```

```python
from pyspark.sql.functions import to_timestamp
```

```python
orders_clean = orders_clean.withColumn("order_timestamp",to_timestamp("order_timestamp","yyyy-MM-dd HH:mm:ss"))
```

```python
orders_clean
```

```python
orders_clean.show()
```

```python
from pyspark.sql.functions import to_date 
```

```python
orders_clean =  orders_clean.withColumn("order_date",to_date("order_timestamp"))
```

```python
orders_clean.show()
```

```python
orders_clean = orders_clean.withColumn("total_amount", col("quantity") * col("price"))
```

```python
orders_clean.show()
```

```python
cust_df.show()
```

```python
from pyspark.sql.functions import upper
```

```python
cust_df = cust_df.withColumn("city", upper(col("city")))
```

```python
cust_df.show()
```

```python
product_df.show()
```

```python
final_join = orders_clean.join(cust_df, on="customer_id", how= "left").join(product_df,on="product_id",how="left")
```

```python
final_join.show()
```

```python
final_df = final_join.select("order_id", 
                  "name", "city", "signup_date", "product_name", "category", 
                  "quantity", "price", "order_timestamp", "order_date", "status", "total_amount" ) 
```

```python
final_df
```

```python
final_df.write.mode("overwrite").parquet("Files/Capstone_1/Output")
```

```python
final_df.write.mode("overwrite").partitionBy("category").parquet("Files/Capstone_1/Output")
```

```python
final_df.show()
```

```python
final_df
```

```python
final_df.agg(sum(col("total_amount")).alias("total_revenue")).show()
```
