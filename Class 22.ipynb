```python
# CountDown
```

```python
class CountDown: 
    def __init__(self, start): 
        self.start = start 

    def __iter__(self):
        return self 

    def __next__(self):
        if self.start <=0: 
            raise StopIteration
        value = self.start 
        self.start = self.start -1 
        return value 
        
```

```python
for i in  CountDown(5): 
    print(i)
```

```python
def countdown(n): 
    while n>0: 
        yield n 
        n = n-1 
```

```python
for i in countdown(5):
    print(i)
```

```python
def decorator(func): 
    def wrapper(): 
        print("Before") 
        func()
        print("After")
    return wrapper  
```

```python
def say_hello(): 
    print("Hello")

say_hello = decorator(say_hello)   #  wrapper  
say_hello() 
```

```python
@decorator
def welcome(): 
    print("Welcome")

#   welcome = decorator(welcome)
```

```python
welcome()
```

```python
def decorator(func): 
    def wrapper(*args,**kwargs): 
        print("Before") 
        res = func(*args,**kwargs)
        print("After")
        return res 
    return wrapper  
```

```python
import time 

def timer(func): 
    def wrapper(*args,**kwargs):
        start = time.time() 
        res = func(*args,**kwargs)
        print(f"Time taken {time.time()-start}")
        return res 
    return wrapper 
```

```python
@timer  
def slow_func(): 
    time.sleep(2)
```

```python
slow_func()
```

```python
def log(func): 
    def wrapper(*args,**kwargs):
        print(f"Running {func.__name__}")
        return func(*args,**kwargs) 
    return wrapper 
```

```python
@log  
def slow_func(): 
    time.sleep(2)
```

```python
slow_func()
```

```python
@timer
@log  
def slow_func(): 
    time.sleep(2)
```

```python
slow_func()
```
