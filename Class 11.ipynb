```python
# nonlocal  keyword
```

```python
def my_out_func():
  x = 300
  def my_in_func():
    nonlocal x
    x = 1000
    print(x)
  my_in_func()
  print(x)
```

```python
my_out_func()
```

```python
def my_out_func():
  x = 300
  def my_in_func():
    #nonlocal x
    x = 1000
    print(x)
  my_in_func()
  print(x)
```

```python
my_out_func()
```

```python
# LEGB Rule
#  Local  -- Inside
#  Enclosing -- Inside enclosing function
#  Global    --  Top level of the module
#  Built-in  --  Built in variables
```

```python
x = 1000
def my_out_func():
  x = 300   # enclosed variable
  def my_in_func():
    print(x)
  my_in_func()
my_out_func()
```

```python
x = 1000
def outer():
  x = 500
  def inner():
    x = 400
    print(x)
  inner()
  print(f"Outer {x}")
outer()
print(f"Global {x}")
```
