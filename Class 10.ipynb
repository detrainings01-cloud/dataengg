```python
def my_func(animal,name):
  print(f"I have an animal {animal} and it's name is {name}")
```

```python
my_func("dog","buddy")   #  Positional Arguments
```

```python
my_func("buddy","dog")
```

```python
my_func(name="buddy",animal="dog")  #kwargs -- Order of assignment doesnot matters
```

```python
def my_func(animal,name, age):
  print(f"I have an animal {animal} and it's name is {name} with {age} years of age")
```

```python
my_func("dog","buddy",5)
```

```python
my_func(name="buddy","dog",5)
```

```python
my_func("dog",5,name="buddy")
```

```python
my_func("dog",name="buddy",age=5)    #     positional args will come first and then kwargs
```

```python
def sum_of(*nums):
  tot_sum = 0
  for i in nums:
    tot_sum = tot_sum + i
  print(tot_sum)
```

```python
def max_value(*nums):
  if len(nums)==0:
    return None
  max_val = nums[0]
  for num in nums:
    if num > max_val:
      max_val = num
  return max_val
```

```python
res = max_value()
```

```python
res
```

```python
res = max_value(1,2,5,4,7,8)
```

```python
res
```

```python
def max_value(*nums):
  if len(nums)==0:
    print("No values passed")
    return None
  max_val = nums[0]
  for num in nums:
    if num > max_val:
      max_val = num
  print(f"max_val is {max_val}")
  return max_val
```

```python
max_value()
```

```python
max_value(1,8,1,41,61)
```

```python
#  kwargs
```

```python
def my_func(**kwargs):
  print(f' firstname is {kwargs.get('fname')}')
  print(f' last name is {kwargs.get('lname')}')
```

```python
my_func(fname="John",lname='D')
```

```python
my_func(fname="John",lname='D',age = 30 )
```

```python
def my_kwargs_func(**kwargs):
  for k,v in kwargs.items():
    print(f"key is {k} and value is {v}")
```

```python
my_kwargs_func(fname="John", lname = "D", age =10, City="Banglore")
```

```python
def my_func(title, *args, **kwargs):
  print(title)
  print("****** Printing args ********")
  for arg in args:
    print(arg)
  print("*** Printing keyword args *****")
  for karg in kwargs:
    print(f' key is {karg} and value is {kwargs[karg]}')
```

```python
my_func("User Info","Abe","Bob","Charan",subject_1="Maths", Subject_2="Science", Subject_3="Social")
```
