```python
my_list = [4,3,2,1]  # List takes more memory  # less speed   #  small datasets
```

```python
def countdown(n): # Generators takes less memory   # faster   #  logs, streaming files
  while n>0:
    yield n
    n = n-1
```

```python
for i in countdown(4):
  print(i)
```

```python
a = 0
b = 1
fibnocci = []  #  0
n = 5
for i in range(n):
  fibnocci.append(a)   #  0
  c = a
  a = b   # 1
  b = c+b  # 2
```

```python
def fibnocci(n):
  a = 0
  b = 1
  fibnocci = []
  for i in range(n):
    fibnocci.append(a)
    a, b = b, a+b
  return fibnocci
```

```python
fibnocci(10)
```

```python
def  fibnocci_gen():
  a,b = 0,1
  while True:
    yield a
    a,b = b, a+b
```

```python
gen = fibnocci_gen()
for i in range(10):
  print(next(gen))
```
