```python
correct_pin = 1234
attempt = 0

while attempt < 3:
  pin= input("Enter pin: ")
  if pin == correct_pin:
    print("Access granted")
    break
  else:
    attempt += 1
    if attempt == 3:
      print("Worng Pin. Card Blocked.")
    else:
      print("Wrong PIN. Please check again.")
```
