```python
employees = {}

while True:
  print("======== Welcome to Employee Management System ========")
  print("1. Add Employee")
  print("2. View Employee")
  print("3. Exit")

  choice = input("Enter your choice (1-3) ")

  if choice=="1":
    emp_id = input("Enter Emp Id ")
    if emp_id in employees:
      print(f"{emp_id} already exists in employees")
    else:
      name = input("Enter Emp Name ")
      role = input("Enter Emp Role ")
      employees[emp_id] = {'name':name, 'role':role}
      print("Added Employee details")

  elif choice == "2":
    print("View Employee")
    if not employees:
      print("No employees")
    else:
      for id,val in employees.items():
          print(f"Emp Id is {id} | name is {val['name']}  |  role is {val['role']}")
  elif choice == "3":
    print("Exiting from Employee Management System")
    break
  else:
    print("Invalid Choice. Enter from 1-3")

```
