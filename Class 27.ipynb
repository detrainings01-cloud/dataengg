```python
employees_df = pd.DataFrame({
    "emp_id": [1, 2, 3, 4, 5],
    "name": ['John', 'Jane', 'Doe', 'Alice', 'Bob'],
    "dept_id": [101, 102, 103, 102, 105],
    "salary": [50000, 60000, 55000, 70000, 65000]
})
```

```python
departments_df = pd.DataFrame({
    "dept_id": [101, 102, 103, 104],
    "department": ['HR', 'Finance', 'IT', 'Marketing']    
})
```

```python
employees_df
```

```python
departments_df
```

```python
pd.merge(employees_df,departments_df, on="dept_id", how="inner")
```

```python
pd.merge(employees_df,departments_df, on="dept_id", how="left")
```

```python
pd.merge(employees_df,departments_df, on="dept_id", how="right")
```

```python
pd.merge(employees_df,departments_df, on="dept_id", how="outer")
```

```python
employees_df = pd.DataFrame({
    "emp_id": [1, 2, 3, 4, 5],
    "name": ['John', 'Jane', 'Doe', 'Alice', 'Bob'],
    "dept_id": [101, 102, 103, 102, 105],
    "salary": [50000, 60000, 55000, 70000, 65000]
})
```

```python
departments_df = pd.DataFrame({
    "id": [101, 102, 103, 104],
    "department": ['HR', 'Finance', 'IT', 'Marketing']    
})
```

```python
pd.merge(employees_df,departments_df, left_on="dept_id", right_on="id",  how="inner") 
```

```python
order_df = pd.DataFrame({"order_id": [1, 2, 3, 4, 5],
                         "customer_id": [101, 102, 103, 104, 105],
                         "order_amount": [1000, 1500, 1200, 2000, 1800]})
```

```python
customer_df = pd.DataFrame({"customer_id": [101, 102, 103, 104, 105],
                            "customer_name": ['Alice', 'Bob', 'Charlie', 'David', 'Eve']})
```

```python
order_df
```

```python
customer_df
```

```python
pd.merge(order_df,customer_df,on="customer_id", how="inner")
```

```python
order_df.join(customer_df.set_index("customer_id"), on="customer_id")
```

```python
sales_df = pd.DataFrame({
    "product_name": ["Laptop", "Shirt", "Sofa", "Smartphone", "Jeans"],
    "Region": ['North', 'South', 'East', 'West', 'North'],
    "Year": [2020, 2020, 2021, 2021, 2022],
    "Sales": [1000, 1500, 1200, 2000, 1800]
})
```

```python
targer_sales_df = pd.DataFrame({
    "product_name": ["Laptop", "Shirt", "Sofa", "Smartphone", "Jeans"],
    "Region": ['North', 'South', 'East', 'West', 'North'],
    "Year": [2020, 2020, 2021, 2021, 2022],
    "Target_Sales": [1100, 1400, 1300, 1900, 1700]
})
```

```python
pd.merge(sales_df,targer_sales_df, on= ["product_name","Region", "Year"], how="inner")
```
