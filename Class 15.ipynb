```python
# List all the files only 
import os

path = "data/"

files = os.listdir(path ) 

for file in files:
    # print(file)
    if os.path.isfile(file):
        print(file)
```

```python
# List all the files only 
import os

path = "data/"

files = os.listdir(path ) 

for file in files:
    file_full_path =  os.path.join(path,file)
    if os.path.isdir(file_full_path):
        print(file)
```
