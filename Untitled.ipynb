```python
import pandas as pd
```

```python
sales_df = pd.DataFrame({
    "product_name": ["Laptop", "Laptop", "Shirt", "Sofa", "Smartphone", "Jeans"],
    "Region": ['North','North', 'South', 'East', 'West', 'North'],
    "Year": [2020,  2020, 2020, 2021, 2021, 2022],
    "Sales": [1000, 1200, 1500, 1200, 2000, 1800]
})
```

```python
targer_sales_df = pd.DataFrame({
    "product_name": ["Laptop", "Shirt", "Sofa", "Smartphone", "Jeans"],
    "Region": ['North', 'South', 'East', 'West', 'North'],
    "Year": [2020, 2020, 2021, 2021, 2022],
    "Target_Sales": [1100, 1400, 1300, 1900, 1700]
})
```

```python
pd.merge(sales_df,targer_sales_df, on= ["product_name","Region", "Year"], how="inner", validate="many_to_one")
```
