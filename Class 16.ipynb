```python
try: 
    num = int(input("Enter the input number: "))
    sqr = num * num 
except ValueError: 
    print("The given input is not a number")
except Exception as e:
    print(e)
else:
    print(sqr)
finally:
    print("Finally block execution")
```

```python
try:   # always 
    num = int(input("Enter the input number: "))
    res = 1/num 
except ValueError:     # On exception 
    print("The given input is not a number")
except Exception as e: # On exception 
    print(e)
else:   #  If no exception 
    print(res)
finally:  # It runs always 
    print("Finally block execution")
```
