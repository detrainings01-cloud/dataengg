```python
class InvalidMarksError(Exception):
    pass
```

```python
# add_student
file_path  = "data/student_db.csv"
def add_student():
    try: 
        sid = int(input("Enter Student Id: "))
        name = input("Enter Student Name: ")
        marks = int(input("Enter Marks: "))

        if marks < 0 or marks > 100:  # not(marks > 0 or marks < 100):
            raise InvalidMarksError(f"Invalid marks entered {marks}. Marks should be between 0 and 100")

        with open(file_path, "a") as file:
            file.write(f"{sid},{name},{marks}\n")
        print("Current record added successfully to file")

    except Exception as e:
        print(f"Error occurred while adding data to student db file. Error is {e}")
        
```

```python
# view student 
def view_student():
    try:
        with open(file_path,"r") as file:
            lines = file.readlines()

        if not lines:
            print("No students data")
        else:
            for line in lines:
                sid,name,marks = line.strip().split(",")
                print(f"ID: {sid} | Name: {name} | Marks: {marks}")
    except Exception as e:
        print(f"Error while executing view_Student {e}")
```

```python
# Search For Student Data
def search_student():
    search_sid = int(input("Enter the Student ID to search: "))
    found = False 
    try:
        with open(file_path,"r") as file:
            for line in file: 
                sid,name,marks = line.strip().split(",")
                if int(sid) == search_sid:  
                    found = True  
                    print(f"ID: {sid} | Name: {name} | Marks: {marks}")
        if not found:
            print(f"Student Id {search_sid} not found in the Student DB")
    except Exception as e:
        print(e)
```

```python
# Delete the student
def del_student():
    try: 
        del_sid = int(input("Enter the student Id to delete: "))
        updated_recs = [] 
        del_id_found = False 
        with open(file_path, "r") as file: 
            for line in file:
                sid,name,marks = line.strip().split(",")
                if int(sid) != del_sid: 
                    updated_recs.append(line)
                else: 
                    del_id_found = True 
                    
        with open(file_path, "w") as file: 
            file.writelines(updated_recs)

        if del_id_found:
            print(f"Deleted student id from the DB ")
        else: 
            print(f"Student ID mentioned is not available in DB")
            
    except Exception as e: 
        print({e})
```

```python
import os
def main():
    if not os.path.exists(file_path):
        with open(file_path,"w"):
            pass  

    while True: 
        print("##########################")
        print("---- Student DB -----")
        print("1.Add Student")
        print("2.View Students")
        print("3.Search Student")
        print("4.Delete Student")
        print("5.Exit")

        choice = int(input("Enter your choice: "))

        if choice == 1: 
            add_student() 
        elif choice == 2:  
            view_student() 
        elif choice == 3:  
            search_student() 
        elif choice == 4:
            del_student() 
        elif choice == 5:
            print("Existing the Application")
            break 
        else: 
            print("Invalid Choice")
```
