```python
class Calculator: 
    def __init__(self, a,b):
        self.a = a 
        self.b = b 
    def add(self): 
        return self.a + self.b 
    def sub(self):
        return self.a - self.b 
    def multiply(self): 
        return self.a * self.b 
    def div(self): 
        return self.a / self.b 
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name   # public 
        self.__age = age  # Private Property   

    def get_age(self):
        return self.__age 

    def set_age(self,age):
        self.__age = age 
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name   # public 
        if self.__validate_age(age):
            self.__age = age  # Private Property   
        else:
            print("Invalid Age")

    def __validate_age(self,age):
        if not isinstance(age,int): 
            return False 
        else:
            return True
    
    def get_age(self):
        return self.__age 

    def set_age(self,age):
        if self.__validate_age(age):
            self.__age = age  # Private Property   
        else:
            print("Invalid Age")
```

```python
p1 = Person("Abe","Eighteen")
```

```python
p1.get_age()
```

```python
p1.__validate_age(18)
```

```python
p1.set_age(18)
```
