```python
def timer(func): 
    def wrapper(*args,**kwargs):
        start = time.time() 
        res = func(*args,**kwargs)
        print(f"function name is {func.__name__} | Time taken {time.time()-start}")
        return res 
    return wrapper 
```

```python
@timer 
def slow_func(): 
    time.sleep(1)
```

```python
slow_func() 
```

```python
def read_logs(file_path): 
    with open(file_path, "r") as file: 
        for line in file: 
            yield line.strip()  
```

```python
# log_pattern 
log_pattern = re.compile(r"(?P<log_time>\d{4}-\d{2}-\d{2}\s\d{2}:\d{2}:\d{2},\d{3}) - "
                        r"(?P<log_level>DEBUG|INFO|WARNING|ERROR|CRITICAL) - "
                        r"(?P<message>.*)"
                        )
```

```python
def log_pattern_func(line): 
    log_match = log_pattern.match(line)
    if log_match: 
        return log_match.groupdict() 
    return None  
```

```python
file_path = "D:\\DE 202512\\CodeFiles\\app.log"
```

```python
for line in read_logs(file_path):
    if log_pattern_func(line): 
        print(log_pattern_func(line)) 
```

```python
parsed_logs = [log_pattern_func(line)   for line in read_logs(file_path)  if log_pattern_func(line) ]
```

```python
parsed_logs
```

```python
for log in parsed_logs:  
    print(log["log_level"])
```

```python
for level  in ["DEBUG","INFO","WARNING","ERROR","CRITICAL"]:
    print(level)
```

```python
@timer 
def analyze_logs(file_path):
    parsed_logs = [log_pattern_func(line)   for line in read_logs(file_path)  if log_pattern_func(line) ]

    summary = { level : sum(1 for log in parsed_logs if log["log_level"] == level)  for level  in ["DEBUG","INFO","WARNING","ERROR","CRITICAL"]  }
    return summary
```

```python
sum_x = analyze_logs(file_path)
```
