```python
customers = {}
```

```python
def create_account(customers):
  acc_no = input("Enter the account number: ")
  if acc_no in customers:
    print(f"{acc_no} already exists")
    return

  name = input("Enter the customer name: ")
  balance = float(input("Enter the initial amount: "))
  customers[acc_no] = {"name":name,"balance":balance}
  print(f"Account added for the name : {name}")
```

```python
def deposit_amount(customers):
  acc_no = input("Enter the account number to deposit: ")
  if acc_no not in customers:
    print(f"Account number does not exists {acc_no}")
    return

  amount = float(input("Enter the amount to deposit: "))
  if amount < 0:
    print("Invalid amount to deposit")
    return

  customers[acc_no]['balance'] += amount   #  customers[acc_no]['balance'] = customers[acc_no]['balance'] + amount
```

```python
def withdraw_amount(customers):
  acc_no = input("Enter the account number to deposit: ")
  if acc_no not in customers:
    print(f"Account number does not exists {acc_no}")
    return

  withdraw_amount = float(input("Enter the amount to withdraw: "))
  if withdraw_amount < 0:
    print("Invalid withdraw amoount")
    return

  balance = customers[acc_no]['balance']
  if balance < withdraw_amount:
    print(f"You do not have enough balance to withdraw")
    return

  customers[acc_no]['balance'] = (balance - withdraw_amount)
  print(f"{withdraw_amount} withdrawn successfully")
```

```python
def check_balance(customers):
  acc_no = input("Enter the account number to deposit: ")
  if acc_no not in customers:
    print(f"Account number does not exists {acc_no}")
    return
  balance = customers[acc_no]['balance']
  print(f"Your account has a balance of {balance}")
```

```python
check_balance(customers)
```

```python
def calc_intrest(customers):
  acc_no = input("Enter the account number to deposit: ")
  if acc_no not in customers:
    print(f"Account number does not exists {acc_no}")
    return
  intrest = lambda bal:bal*0.05
  intrest_earned = intrest(customers[acc_no]['balance'])
  customers[acc_no]['balance'] += intrest_earned
  print(f'Account earned an intrest of {intrest_earned} and updated balance is {customers[acc_no]['balance']}')
```

```python
def bank_application():
  while True:
    print("***********************************************************")
    print("Welcome to bank application. Enter your choice between 1-6")
    print("1. Create account")
    print("2. Deposit amount")
    print("3. Withdraw amount")
    print("4. Check balance")
    print("5. Calculate intrest")
    print("6. Exit")
    choice = int(input("Enter choice (1-6): "))
    if choice == 1:
      create_account(customers)
    elif choice == 2:
      deposit_amount(customers)
    elif choice == 3:
      withdraw_amount(customers)
    elif choice == 4:
      check_balance(customers)
    elif choice == 5:
      calc_intrest(customers)
    elif choice == 6:
      print(f'Exiting from the Bank Application')
      break
    else:
      print("Enter the valid choice between 1-6")
```
