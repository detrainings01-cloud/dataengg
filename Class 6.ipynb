```python
my_nums = [10,20,30,40]
for i in my_nums:
  print(i*2)
```

```python
my_nums = [10,20,30,40]
res = []
for i in my_nums:
  res.append(i*2)
print(res)
```

```python
[ i*2 for i in my_nums]
```

```python
fruit_names = ['apple','banana','cherry','orange','mango']
new_fruits = []

for i in fruit_names:
  if len(i) > 5:
    new_fruits.append(i)
print(new_fruits)
```
