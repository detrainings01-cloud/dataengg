```python
import random
score = 0
for i in range(1,6):
  a = random.randint(1,10)
  b = random.randint(1,10)

  answer = int(input(f'Enter {a} * {b}:'))

  res = a*b
  if res == answer:
    print(f"You entered corred value {a}*{b}={res}")
    score = score + 1
  else:
    print(f"Wrong answer {a}*{b}={res}; You entered {answer}")

print(f"You got a score of {score}")
```
