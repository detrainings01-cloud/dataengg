```python
from abc import ABC, abstractmethod 
class Vehicle(ABC): 
    @abstractmethod 
    def start(self):
        pass 
```

```python
v = Vehicle() 
```

```python
class Car(Vehicle): 
    def start(self): 
        print("Car starts with a Key")
```

```python
c = Car() 
c.start() 
```

```python
from abc import ABC, abstractmethod 
class Payment(ABC):
    @abstractmethod
    def pay(self,amount): 
        pass

class UPI(Payment): 
    def pay(self, amount): 
        print(f"Payment from UPI and the amount is {amount}")
    
class CreditCard(Payment):
    def pay(self, amount): 
        print(f"Payment from Credit card and the amount is {amount}")
```

```python
def process_payment(payment: Payment, amount): 
    payment.pay(amount)
```

```python
process_payment(UPI(),100)
```

```python
process_payment(CreditCard(),1000)
```

```python
class Book: 
    def __init__(self,book_id, title, author, is_availabe): 
        self.book_id = book_id 
        self.title = title  
        self.author = author 
        self.is_availabe = is_availabe  

    def display(self) : 
        print(f"{self.book_id} | {self.title} | {self.author} | {self.is_availabe}")
```

```python
b1 = Book(1,"Python","Abe",True)
```

```python
b1.display() 
```

```python
class User: 
    def __init__(self,user_id,name): 
        self.user_id = user_id 
        self.name = name  
        self.borrowed_books = [] 

    def borrow_book(self,book): 
        self.borrowed_books.append(book) 

    def return_book(self,book): 
        self.borrowed_books.remove(book) 

    def view_borrowed_books(self): 
        if not self.borrowed_books: 
            print("No books borrowed")
            return 
        for  book in self.borrowed_books: 
            print(book) 
```

```python
u1 = User(1,"User1")
```

```python
class Library: 
    def __init__(self): 
        self.books = [] 
        self.users = [] 

    def add_user(self): 
        name = input("Enter user name" )
        uid  = input("Enter user Id")

        for user in self.users: 
            if user.user_id == uid : 
                print("User Id already exists")
                return 

        u = User(uid,name)
        self.users.append(u)
        print("User added successfully")

    def user_data(self):
        for user in self.users: 
            print(f"{user.user_id} | {user.name} | {user.borrowed_books}") 

    def add_books(self): 
        book_id = input("Enter the book_id")
        title = input("Enter the title")
        author = input("Enter the author")

        for book in self.books: 
            if book.book_id == book_id: 
                print("Book Id already exists") 
                return 

        b = Book(book_id,title,author,True)
        self.books.append(b)
        print("Book details entered")

    def  view_books(self): 
        if not self.books: 
            print("No books available")
            return 
        for book in self.books: 
            book.display() 
    
    def get_user(self,user_id): 
        for user in self.users:  
            if user.user_id == user_id:
                return user 
        return None 

    def get_book(self,book_id): 
        for book in self.books:  
            if book.book_id == book_id:
                return book 
        return None 

    def borrow_book(self): 
        user_id = input("Enter user id ")
        user = self.get_user(user_id)
        if not user:
            print("User not available ")
            return 
            
        book_id = input("Enter Book id ") 
        book = self.get_book(book_id) 
        if not book: 
            print("Book not available")
            return  

        if not book.is_availabe:
            print("Book is not available")
            return 

        book.is_availabe = False 
        user.borrow_book(book)
        print("Book borrowed successfully")

    def return_book(self):
        user_id = input("Enter user id ")
        book_id = input("Enter Book id ") 
        
        user = self.get_user(user_id) 
        book = self.get_book(book_id) 

        if not user or not book: 
            print("Invalid User ID or Book Id")
            return 

        if book not in user.borrowed_books: 
            print("This is not borrowed by the user")
            return 

        book.is_availabe = True 
        user.return_book(book)
```
