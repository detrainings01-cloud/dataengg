```python
data = [
    (1, "Abe", "Laptop", 80000, "Electronics", "2025-01-01"),
    (2, "Maya", "Smartphone", 45000, "Electronics", "2025-01-05"),
    (3, "John", "Refrigerator", 60000, "Home Appliances", "2025-01-10"),
    (4, "Sara", "Washing Machine", 35000, "Home Appliances", "2025-01-12"),
    (5, "Abe", "Headphones", 5000, "Electronics", "2025-01-15"),
]
columns = ["order_id","Name","product","amount","category","order_date"] 
df = spark.createDataFrame(data,columns)
```

```python
df.groupBy("category").agg(sum("amount").alias("total_revenue")).show()
```

```python
df.groupBy("category").agg(avg("amount").alias("avg_revenue")).show()
```

```python
df.groupBy("category").agg(count("order_id").alias("count_orders")).show()
```

```python
df.groupBy("category").agg(sum("amount").alias("total_revenue"), avg("amount").alias("avg_revenue"), count("order_id").alias("count_orders")).show()
```

```python
df.groupBy("category")\
    .agg(sum("amount").alias("total_revenue"), 
         avg("amount").alias("avg_revenue"), 
         count("order_id").alias("count_orders")).show()
```

```python
df.createOrReplaceTempView("orders")
```

```python
query = "select category, sum(amount) as  total_revenue, avg(amount) as avg_revenue, count(order_id) as count_orders from orders group by category "
```

```python
customer_data = [("Abe","Hyderabad"), ("Maya","Bangalore"), ("John","Chennai")]
cust_columns = ["Name","City"]
```

```python
df_cust = spark.createDataFrame(customer_data,cust_columns)
```

```python
df_cust.show()
```

```python
df_joined = df.join(df_cust,on="Name",how="inner")
```

```python
df_joined.show()
```

```python
df_joined = df.join(df_cust,on="Name",how="left")  # how = inner, left, right, outer 
```

```python
df_joined.show()
```
